```python
from sagemaker import get_execution_role
from sagemaker.pytorch.estimator import PyTorch

role = get_execution_role()
```

```python
estimator = PyTorch(
    source_dir='./code/',
    entry_point='HetGNN.py',
    framework_version='1.10',
    py_version='py38',
    instance_type='ml.p2.xlarge',
    instance_count=1,
    volume_size=100,
    role=role,
    base_job_name='HetGNN'
)
```

```python
estimator.fit(
    {
        'train': 's3://prod-tpgt-knowledge-lake-sandpit-v1/application/anomaly_detection/deeptralog/HetGNN/ProcessedData_rw_top10/feature_list/'
    }
)
```
